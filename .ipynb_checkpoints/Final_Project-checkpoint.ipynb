```python
analysis = ratings[['ShootingTotal', 'PassingTotal', 'PaceTotal', 'DefendingTotal', 
                    'DribblingTotal', 'PhysicalityTotal', 'BestPosition']]
```

```python
analysis.describe()
```

```python
mean = analysis.ShootingTotal.mean()
STD = analysis.ShootingTotal.std()
Max = analysis.ShootingTotal.max()
Min = analysis.ShootingTotal.min()
common = analysis.ShootingTotal.mode()
```

```python
shooting = {
    'Average': [mean],
    'STD': [STD],
    'Max': [Max],
    'Min': [Min],
    'Common': [common]
}
shooting_df = pd.DataFrame(shooting)
shooting_df

shooting_df = shooting_df.reset_index().melt(var_name='Stat', value_name='Value')
shooting_df = shooting_df.set_index('Stat')
shooting_df
```

```python
# passing

mean = analysis.PassingTotal.mean()
STD = analysis.PassingTotal.std()
Max = analysis.PassingTotal.max()
Min = analysis.PassingTotal.min()
common = analysis.PassingTotal.mode()

passing = {
    'Average': [mean],
    'STD': [STD],
    'Max': [Max],
    'Min': [Min],
    'Common': [common]
}
passing_df = pd.DataFrame(passing)

passing_df = passing_df.reset_index().melt(var_name='Stat', value_name='Value')
passing_df = passing_df.set_index('Stat')
passing_df
```

```python
# pace

mean = analysis.PaceTotal.mean()
STD = analysis.PaceTotal.std()
Max = analysis.PaceTotal.max()
Min = analysis.PaceTotal.min()
common = analysis.PaceTotal.mode()

pace = {
    'Average': [mean],
    'STD': [STD],
    'Max': [Max],
    'Min': [Min],
    'Common': [common]
}
pace_df = pd.DataFrame(pace)

pace_df = pace_df.reset_index().melt(var_name='Stat', value_name='Value')
pace_df = pace_df.set_index('Stat')
pace_df
```

```python
# dribbling

mean = analysis.DribblingTotal.mean()
STD = analysis.DribblingTotal.std()
Max = analysis.DribblingTotal.max()
Min = analysis.DribblingTotal.min()
common = analysis.DribblingTotal.mode()

dribbling = {
    'Average': [mean],
    'STD': [STD],
    'Max': [Max],
    'Min': [Min],
    'Common': [common]
}
dribbling_df = pd.DataFrame(dribbling)

dribbling_df = dribbling_df.reset_index().melt(var_name='Stat', value_name='Value')
dribbling_df = dribbling_df.set_index('Stat')
dribbling_df
```

```python
features = ['PaceTotal', 'ShootingTotal', 'PassingTotal', 'DribblingTotal', 'DefendingTotal', 'PhysicalityTotal']

for feature in features:
    fig, axs = plt.subplots(ncols=2, figsize=(10,4))
    
    sns.histplot(data=analysis, x=feature, kde=False, ax=axs[0], bins=10)
    axs[0].set_title('Histogram of ' + feature)
    
    sns.kdeplot(data=analysis, x=feature, ax=axs[1])
    axs[1].set_title('Density plot of ' + feature)
    
    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_avg_ratings_by_position(analysis):
    colors = ['r', 'b', 'g', 'y', 'm', 'c']
    avg_ratings = analysis.groupby('BestPosition')[['PaceTotal', 'ShootingTotal', 'PassingTotal', 'DribblingTotal', 'DefendingTotal', 'PhysicalityTotal']].mean()
    
    for position in avg_ratings.index:
        fig, ax = plt.subplots(figsize=(6,4))
        
        avg_ratings.loc[position].plot(kind='bar', ax=ax)
        
        ax.set_title('Average Ratings for {}'.format(position))
        ax.set_xlabel('Attribute')
        ax.set_ylabel('Average Rating')
        
        plt.show()

```

```python
plot_avg_ratings_by_position(analysis)
```
